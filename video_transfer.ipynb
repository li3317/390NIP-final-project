```
import cv2
import numpy as np
import os
from PIL import Image, ImageEnhance

# Input and Output path:
input = '/content/drive/MyDrive/Video Transfer/Chess club.mp4'
output = '/content/drive/MyDrive/Video Transfer/Frames/'
HEIGHT = 256
WIDTH = 256

# Playing video from file:
cap = cv2.VideoCapture(input)

try:
    if not os.path.exists('data'):
        os.makedirs('data')
except OSError:
    print ('Error: Creating directory of data')

currentFrame = 0
while(True):
    # Capture frame-by-frame
    ret, frame = cap.read()
    if not ret: 
      break

    # Saves image of the current frame in jpg file
    # Crop face
    crop = frame[0:700, 0:700]
    # Resize
    resize = cv2.resize(crop, (HEIGHT, WIDTH))
    # Save
    cv2.imwrite(output + 'frame%d.jpg' % currentFrame, resize)

    print ('Creating...' + output)

    # To stop duplicate images
    currentFrame += 1

# When everything done, release the capture
cap.release()
cv2.destroyAllWindows()
```

```
import os
import moviepy.video.io.ImageSequenceClip
from moviepy.editor import *
from PIL import Image, ImageEnhance
from natsort import natsorted

# Folder location
image_folder = '/content/drive/MyDrive/Video Transfer/Edge_Frames'
# Output name
video_name = 'final.mp4'
# Frame rate
fps=30


# Audio Extract
video = VideoFileClip(input)
audio = video.audio
audio.write_audiofile('/content/drive/MyDrive/Video Transfer/og_voice.mp3')

# Extract Image and construct video
image_files = [image_folder +'/'+ img for img in os.listdir(image_folder) if img.endswith(".png")]

# Reembed Audio
audioclip = AudioFileClip('/content/drive/MyDrive/Video Transfer/og_voice.mp3')

clip = moviepy.video.io.ImageSequenceClip.ImageSequenceClip(natsorted(image_files), fps=fps)
new_audioclip = CompositeAudioClip([audioclip])
clip.audio = new_audioclip
clip.write_videofile('/content/drive/MyDrive/' + video_name)
```
